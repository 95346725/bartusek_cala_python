```python
#stock and indices dictionary
import pandas as pd
import re
import json
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
scrapedSAP = data[0]
tickers = scrapedSAP['Symbol'].tolist()
fullName = scrapedSAP['Security'].tolist()
#print(fullName)
stockDict = dict()
for i in range(len(tickers)):
   stockDict[tickers[i]] = [tickers[i],fullName[i]]

unwantedSuffix = [' & Co\.$',
                  ' Co\.$',
                  ' Company$',
                  ' Ltd$',
                  ' Ltd\.$',
                  ' Inc\.$',
                  ', Inc$',
                  ' Corp\.$',
                  ' Corp$',
                  ' Inc$',
                  ' Co$',
                  ' Corporation$',
                  "\.com$"]

for key in stockDict:
    stockDict[key].append('$'+stockDict[key][0])
    for j in unwantedSuffix:
        name = stockDict[key][1]
        stockDict[key][1] = re.sub(j,'',name)
        
        
simpleTickers = []
for key in stockDict:
    simpleTickers.append(len(key) == 1)

#remove one letter tickers
counter = 0   
test = stockDict.copy() #cant change dictionary during iteration, so we iterate over a copy of this dictionary
for key in stockDict.copy():
    if simpleTickers[counter]:
        stockDict.pop(key)
    counter +=1
#expand for selected indexes
stockDict['SPGI'][1] = 'S&P'
stockDict['SPGI'].append("S&P500")
stockDict['NDAQ'].append("NASDAQ100")
stockDict['DOW'].append("DJIA")
stockDict['DOW'].append("DJI")
stockDict['ARKK'] = ['ARKK'] #because ark funds are popular nowadays
#manually remove stocks which have tickers similar to common english words or could likely be matched by mistake
stockDict.pop('ARE')
stockDict.pop('ALL')
stockDict.pop('CAT')
stockDict.pop('COST')
stockDict.pop('DD') #due diligence
stockDict.pop('FAST')
stockDict.pop('IT')
stockDict.pop('PEAK')
stockDict.pop('INFO')
stockDict.pop('KEY')
stockDict.pop('KEYS')
stockDict.pop('LOW')
stockDict.pop('TAP')
stockDict.pop('POOL')
stockDict.pop('NOW')
stockDict.pop('SO')
stockDict.pop('SEE')
```

```python
#dump the dictionary as json for use in other 
dictJson = open("stockDict.json", "w")
json.dump(stockDict, dictJson)
dictJson.close()
```
